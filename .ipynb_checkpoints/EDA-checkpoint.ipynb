## Import & functions

```python
import os
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def show_images(sign_class, url, n_rows=2, n_cols=5):
    '''
    Function for visualization images for selected traffic sign class
    sign_class: name of folder with one traffic sign class
    url: path to the "sign_class" folder
    n_rows: number of image rows
    n_cols: number of images in each row
    '''
    img_path = os.path.join(url, sign_class)
    images_list = os.listdir(img_path)
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(6*n_cols, 6*n_rows))
    i, j = 0, 0
    print(f'Random {n_rows*n_cols} images (for sign_class={sign_class}):')
    for n in range(n_rows*n_cols):
        pic = np.random.choice(images_list)
        address = os.path.join(img_path, pic)
        img = mpimg.imread(address)
        axes[j][i].imshow(img)
        axes[j][i].set_title(pic)
        i += 1
        j += i // n_cols
        i %= n_cols
    plt.figure(figsize=(10, 16))
    plt.show()
```

## Visualization of types of traffic signs

```python
# Random traffic signs (one type of sign - one "sign_class")
url_meta = 'E:\DataSets\Traffic Sign - Detection&Recognation\Traffic_Sign - 200 classes'
show_images('Meta', url_meta, n_rows=6, n_cols=6)
```

## Analysis of data on the number of images for classes

```python
url_train = r'E:\DataSets\Traffic Sign - Detection&Recognation\Traffic_Sign - 200 classes\Train'
n_classes = len(os.listdir(url_train))
print(f'Check number of sign classes: {n_classes}')
```

```python
# Create dictionary with quantities for each class
train_class_dict = {}
for dir in os.listdir(url_train):
    dir_path = os.path.join(url_train, dir)
    train_class_dict[int(dir)] = len(os.listdir(dir_path))

# Creat DataFrame
df = pd.DataFrame.from_dict(train_class_dict, orient='index', columns=['train'])
df.sort_index(inplace=True)
print(f'Total namber of images: {df.train.sum()} (average: {round(df.train.sum()/n_classes)} images per class)')
print(f'Class with minimum quantities images: {df.index[df.train==df.train.min()].values} class - {df.train.min()} images')
print(f'Class with maximum quantities images: {df.index[df.train==df.train.max()].values} class - {df.train.max()} images')

# Plot with number of images for each class in Train Data (except zero class)
ax = df.iloc[1:, :].plot(kind='bar', figsize=(12, 4), title='Number of images for each class in Train Data')
ax.set_xticklabels([])
ax.set_xlabel('Classes: 1 -> 199 (excluding class-0)')
plt.show()
```

```python
# Statistical distribution of the number of images by class
df.train.describe()
```

## Analisis of data by size images

```python
# Create list with minimum shape for all images (image: a*b -> min(a, b))
img_sizes = []
for dir_name in os.listdir(url_train):
    url_dir = os.path.join(url_train, dir_name)
    for file in os.listdir(url_dir):
        url_img = os.path.join(url_dir, file)
        img = mpimg.imread(url_img)
        img_shape = img.shape[:2]
        img_sizes.append(img_shape)
```

```python
# Statistical distribution of the number of image shapes
shape_images_df = pd.DataFrame(img_sizes, columns=['height', 'width'])
shape_images_df['min_size'] = shape_images_df[['height', 'width']].min(axis=1)
shape_images_df.describe()
```

```python
# Plot with number of images for Top-100 min_sizes
shape_df = shape_images_df['min_size'].value_counts().sort_index()
shape_df.head(120).plot(kind='bar', figsize=(12, 4), title='Number of images for Top-100 min_sizes (Train Data)', fontsize=6)
```

#### According to the chart above, we can choose the minimum size - 30 pixels to start training the model
